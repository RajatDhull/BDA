```python
import pandas as pd
import numpy as np 

df = pd.read_csv('selected_data1.csv')
df.info()
```

```python
data = df['SrcAddr']
print(data)
```

```python
import mmh3  # MurmurHash3 hash function

window_size = 100
hash_function = lambda x : mmh3.hash(str(x), seed=0)
```

```python
def find_trailing_zeroes(element):
    hash_value = hash_function(element)
    hash_value = abs(hash_value)%(100)
    # print(hash_value)
    binary = bin(hash_value)[2:]


    # Count the number of trailing zeros
    trailing_zeroes = len(binary) - len(binary.rstrip('0'))
    return trailing_zeroes

distinct_count = []
for i in range(0, min(10000,len(data) - window_size + 1)):
    window = data[i:i+window_size]
    R = 0
    for element in window:
        r = find_trailing_zeroes(element)
        # print("r", r)
        R = max(r, R)
    
    # print(R)
    distinct_counts = pow(2, R)
    distinct_count.append(distinct_counts)

    print(f"Window {i+1}-{i+window_size}:Predicted Distinct Counts - {distinct_counts}")
```

```python
threshold = 50
attack_windows = []
for i in range(0,len(distinct_count)):
    if distinct_count[i] > threshold:
        attack_windows.append((i+1, i + window_size))
#         print(distinct_count[i])
        
print("Detected Attack Windows:")
# print(attack_windows)
# print(len(attack_windows))
for start, end in attack_windows:
    print(f"Start Index: {start}, End Index: {end}")
```

```python
window_size = 100

# List to store distinct elements in each window
distinct_elements = []
true_attack_windows = []
# Iterate through the dataset with a sliding window approach
for i in range(0, len(data) - window_size + 1):
    window = data[i:i + window_size]  # Change 'Column_Name' to the actual column name in your dataset
    
    # Get distinct elements in the current window
    distinct_elements_window = window.unique().tolist()
    
    # Add distinct elements to the list
    distinct_elements.append(len(distinct_elements_window))
    

for i in range(0,len(distinct_elements)):
    if distinct_elements[i] > threshold:
        true_attack_windows.append((i+1, i + window_size))
# Print distinct elements in each window
for i in range(0,len(distinct_elements)):
    print(f"Window {i+1}-{i+window_size}: True Distinct Counts - {distinct_elements[i]}")
```

```python
print("True Attack Windows:")
for start, end in true_attack_windows:
    print(f"Start Index: {start}, End Index: {end}")
```

```python
accurate_count = 0;
for i in range(0,len(distinct_elements)):
    if distinct_elements[i] == distinct_count[i]:
        accurate_count = accurate_count + 1;
        
accuracy = accurate_count/(len(distinct_elements)) * 100


accurate_detection = 0
# print(attack_windows)
for i in range(0,len(true_attack_windows)):
#     print(true_attack_windows)
    if true_attack_windows[i] in attack_windows:
        accurate_detection = accurate_detection + 1

detection_accuracy = accurate_detection/(len(true_attack_windows)) * 100
print("Accuracy of detection: ",detection_accuracy)
print("Accuracy of Counting distinct elements: ",accuracy)
```
